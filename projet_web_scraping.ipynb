```python
import requests
from bs4 import BeautifulSoup
import csv
from itertools import zip_longest
```

```python
nom_articles = []
articles_prices_mytek = []
articles_prices_tunisianet = []
description_list = []
ecran_size_list = []
type_processeur_list = []
ram_list = []
rom_list = []
carte_graphique_list = [] 

sys_exploitation_list = []
ref_processeur_list = []
ecran_type_list = []
graphic_card_ref_list = []
garanti_list =[]
article_link = []
store = []
article_reference = []
image = []
```

SCRAPING MYTECK

```python
page_number = 1
mytek_link_list =[]
while True :
    result = requests.get(f"https://www.mytek.tn/informatique/ordinateurs-portables/pc-portable.html?p={page_number}")    
    src = result.content
    soup = BeautifulSoup(src,"lxml")
    reference = soup.find_all("div",{"class":"sku"})
    
    #images = soup.find_all("span",{"class":"product-image-wrapper"})
    for img in soup.find_all("img",{"class":"product-image-photo"}) :
        image.append( img.get('src') )


    for i in range(len(reference)):
        article_reference.append(reference[i].text)
    nom_article = soup.find_all("strong",{"class":"product name product-item-name"})
    for i in range(len(nom_article)):
        #if (nom_article[i].text not in nom_articles) :
        mytek_link_list.append(nom_article[i].find("a").attrs['href'])
        nom_articles.append(nom_article[i].text)
    prix_article = soup.find_all("span",{"data-price-type":"finalPrice","class":"price-wrapper"})
    for i in range(len(prix_article)):
        articles_prices_mytek.append(prix_article[i].text)
    #for i in range(len(articles_prices),-1,-1) :
    #    if (i % 2 != 0) :
    #        articles_prices.pop(i)
    print("scraping page number "+ str(page_number))
    if (page_number == 100 ) :
        break    
    page_number += 1
```

```python
for i in range(len(articles_prices_mytek),-1,-1) :
        if (i % 2 != 0) :
            articles_prices_mytek.pop(i)

for i in range(len(nom_articles),-1,-1) :
        if (i % 2 != 0) :
            nom_articles.pop(i)
            mytek_link_list.pop(i)
```

```python
for link in mytek_link_list :
    result = requests.get(link)
    src = result.content
    soup = BeautifulSoup(src,"lxml")
     
    taille_ecran = soup.find("td",{ "class":"col data","data-th":"Taille de l'écran"})   
    type_processeur = soup.find("td",{ "class":"col data","data-th":"Type Processeur"})
    ram = soup.find("td",{ "class":"col data","data-th":"Mémoire"})
    rom = soup.find("td",{ "class":"col data","data-th":"Disque Dur"})
    carte_graphique = soup.find("td",{ "class":"col data","data-th":"Carte Graphique"})
    sys_exploitation = soup.find("td",{ "class":"col data","data-th":"Système d'exploitation"})
    ref_processeur = soup.find("td",{ "class":"col data","data-th":"Fréquence Processeur"})
    ecran_type = soup.find("td",{ "class":"col data","data-th":"Ecran"})
    graphic_card_ref = soup.find("td",{ "class":"col data","data-th":"Chipset Graphique"})
    garanti = soup.find("td",{ "class":"col data","data-th":"Garantie"})

    ecran_size_list.append(taille_ecran.text)
    type_processeur_list.append(type_processeur.text)
    ram_list.append(ram.text)
    rom_list.append(rom.text)
    carte_graphique_list.append(carte_graphique.text)
    sys_exploitation_list.append(sys_exploitation.text)
    ref_processeur_list.append(ref_processeur.text)
    ecran_type_list.append(ecran_type.text)
    graphic_card_ref_list.append(graphic_card_ref.text)
    garanti_list.append(garanti.text)
    article_link.append(link)
    store.append('Mytek')
    

```

SCRAPING TUNISIANET

```python
import urllib3
urllib3.disable_warnings()
```

```python
page_number = 1
products=[]
tunisianet_links = []
while True :
    result = requests.get(f"https://www.tunisianet.com.tn/301-pc-portable-tunisie?page={page_number}" , verify=False)    
    src = result.content
    soup = BeautifulSoup(src,"lxml")
    reference = soup.find_all("span",{"class":"product-reference"})
 
    for img in soup.find_all("img",{"class":"second-img img-responsive center-block"}) :
        image.append( img.get('src') )

    for i in range(len(reference)):
        x = reference[i].text
        x=x.replace("[","")
        x=x.replace("]","")
        article_reference.append(x)
    products = soup.find_all("div",{"class":"item-product col-xs-12"})
    for i in range(len(products)):
        tunisianet_links.append(products[i].find("a").attrs['href'])
        nom_article=products[i].find("h2")
        nom_articles.append(nom_article.text)
    prix_article = soup.find_all("span",{"itemprop":"price","class":"price"})
    for i in range(len(prix_article)):
        articles_prices_tunisianet.append(prix_article[i].text)
    print("scraping page number"+str(page_number))
    if (page_number == 40 ) :
        break    
    page_number += 1
```

```python
for i in range(len(articles_prices_tunisianet),-1,-1) :
        if (i % 2 != 0) :
            articles_prices_tunisianet.pop(i)
```

```python
for link in tunisianet_links :
    article_link.append(link)
    store.append("Tunisianet")
    result = requests.get(link,verify=False)
    src = result.content
    soup = BeautifulSoup(src,"lxml")
    d = soup.find("dl",{"class":"data-sheet"})
    dt = d.find_all("dt") 
    dd = d.find_all("dd")
    for i in range(len(dt)):
        if (dt[i].text == "Disque Dur" ) :
            rom_list.append(dd[i].text)
        elif (dt[i].text == "Carte Graphique") :
            carte_graphique_list.append(dd[i].text)
        elif (dt[i].text == "Taille Ecran") :
            ecran_size_list.append(dd[i].text)
        elif (dt[i].text == "Processeur") :
            type_processeur_list.append(dd[i].text)
        elif (dt[i].text == "Mémoire") :
            ram_list.append(dd[i].text)
        elif (dt[i].text == "Système d'exploitation") :
            sys_exploitation_list.append(dd[i].text)
        elif (dt[i].text == "Réf processeur") :
            ref_processeur_list.append(dd[i].text)
        elif (dt[i].text == "Type Ecran") :
            ecran_type_list.append(dd[i].text)
        elif (dt[i].text == "Réf Carte graphique") :
            graphic_card_ref_list.append(dd[i].text)
        elif (dt[i].text == "Garantie ") :
            garanti_list.append(dd[i].text)

    x = max(len(rom_list),len(carte_graphique_list),len(ecran_size_list),len(type_processeur_list),len(ram_list),len(sys_exploitation_list),len(ref_processeur_list),len(ecran_type_list),len(graphic_card_ref_list),len(garanti_list))     
    #print(x)
    if (len(rom_list) < x):
        rom_list.append(None)
    if (len(carte_graphique_list) < x):
        carte_graphique_list.append(None)
    if (len(ecran_size_list) < x):
        ecran_size_list.append(None)
    if (len(type_processeur_list) < x):
        type_processeur_list.append(None)
    if (len(ram_list) < x):
        ram_list.append(None)
    if (len(sys_exploitation_list) < x):
        sys_exploitation_list.append(None)
    if (len(ref_processeur_list) < x):
        ref_processeur_list.append(None)
    if (len(ecran_type_list) < x):
        ecran_type_list.append(None)    
    if (len(graphic_card_ref_list) < x):
        graphic_card_ref_list.append(None)
    if (len(garanti_list) < x):
        garanti_list.append(None)      

    """if len(rom_list) !=  len(carte_graphique_list) or len(rom_list) != len(ecran_size_list) or len(rom_list) !=len(type_processeur_list) or len(rom_list) !=len(ram_list) or len(rom_list) !=len(sys_exploitation_list) or len(rom_list) != len(ref_processeur_list) or len(rom_list) !=len(ecran_type_list) or len(rom_list) != len(graphic_card_ref_list) or len(rom_list) != len(garanti_list) :
        

        print(dt)
        print(dd)
        
        print(len(rom_list))
        print(len(carte_graphique_list))
        print(len(ecran_size_list))
        print(len(type_processeur_list))
        print(len(sys_exploitation_list))
        print(len(ref_processeur_list))
        print(len(ecran_type_list))
        print(len(graphic_card_ref_list))
        print(len(garanti_list))
        break"""
            
```
