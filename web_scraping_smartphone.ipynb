```python
import requests
from bs4 import BeautifulSoup
import csv
from itertools import zip_longest
import pandas as pd
import pymysql
import sqlite3
import pyodbc
```

```python
nom_articles = []
article_reference = []
articles_prices_mytek = []
articles_prices_tunisianet = []

ecran_size_list = []
type_processeur_list = []
ram_list = []
rom_list = []
sys_exploitation_list = []
camera_list = []

article_link = []
store = []
article_reference = []
image = []
```

```python
page_number = 1
mytek_link_list =[]

articles_prices_mytek = []

while True :
    result = requests.get(f"https://www.mytek.tn/telephonie-tunisie/smartphone-mobile-tunisie/smartphone-tunisie.html?p={page_number}")    
    src = result.content
    soup = BeautifulSoup(src,"lxml")
    reference = soup.find_all("div",{"class":"sku"})
    
    #images = soup.find_all("span",{"class":"product-image-wrapper"})
    for img in soup.find_all("img",{"class":"product-image-photo"}) :
        if img == None :
            image.append( None )
        else :
            image.append( img.get('src') )


    for i in range(len(reference)):
        article_reference.append(reference[i].text)
    nom_article = soup.find_all("strong",{"class":"product name product-item-name"})
    for i in range(len(nom_article)):
        #if (nom_article[i].text not in nom_articles) :
        mytek_link_list.append(nom_article[i].find("a").attrs['href'])
        nom_articles.append(nom_article[i].text)
    prix_article = soup.find_all("span",{"data-price-type":"finalPrice","class":"price-wrapper"})
    for i in range(len(prix_article)):
        articles_prices_mytek.append(prix_article[i].text)
    #for i in range(len(articles_prices),-1,-1) :
    #    if (i % 2 != 0) :
    #        articles_prices.pop(i)
    print("scraping page number "+ str(page_number))
    if (page_number == 42 ) :
        break    
    page_number += 1
```

```python
for link in mytek_link_list :
    result = requests.get(link)
    src = result.content
    soup = BeautifulSoup(src,"lxml")
     
     
    ecran_size = soup.find("td",{ "class":"col data","data-th":"Taille de l'écran"})
    ram = soup.find("td",{ "class":"col data","data-th":"Mémoire"})
    rom = soup.find("td",{ "class":"col data","data-th":"Stockage"})
    sys_exploitation = soup.find("td",{ "class":"col data","data-th":"Système d'exploitation"})
    type_processeur = soup.find("td",{ "class":"col data","data-th":"Type Processeur"})
    camera = soup.find("td",{ "class":"col data","data-th":"Appareil Photo"})
   
  

    
    if ecran_size == None :
        ecran_size_list.append(None)  
    else :
        ecran_size_list.append(ecran_size.text)

    if ram == None :
        ram_list.append(None)
    else :
        ram_list.append(ram.text)

    if rom == None :
        rom_list.append(None)
    else :
        rom_list.append(rom.text)

    sys_exploitation_list.append(sys_exploitation.text)

    if type_processeur == None :
        type_processeur_list.append(None)
    else :
        type_processeur_list.append(type_processeur.text)

    if camera == None :
        camera_list.append(None)
    else :
        camera_list.append(camera.text)
   
    article_link.append(link)
    store.append('Mytek')

    

```

```python
page_number = 1
products=[]
tunisianet_links = []

while True :
    v = 0
    result = requests.get(f"https://www.tunisianet.com.tn/596-smartphone-mobile-4g-tunisie?page={page_number}" , verify=False)    
    src = result.content
    soup = BeautifulSoup(src,"lxml")
    reference = soup.find_all("span",{"class":"product-reference"})
 
    l = soup.find_all("img",{"class":"second-img img-responsive center-block"})
    
    for img in l :
        if img == None :
            image.append(None)
        else :
            image.append( img.get('src') )
        v=v+1

    for i in range(len(reference)):
        x = reference[i].text
        x=x.replace("[","")
        x=x.replace("]","")
        article_reference.append(x)
    products = soup.find_all("div",{"class":"item-product col-xs-12"})
    for i in range(len(products)):
        tunisianet_links.append(products[i].find("a").attrs['href'])
        nom_article=products[i].find("h2")
        nom_articles.append(nom_article.text)
    prix_article = soup.find_all("span",{"itemprop":"price","class":"price"})
    for i in range(len(prix_article)):
        articles_prices_tunisianet.append(prix_article[i].text)
    print("scraping page number"+str(page_number))
    if v < 24 :
        image.append(l[len(l)-1])
    if (page_number == 8 ) :
        break    
    page_number += 1
```
